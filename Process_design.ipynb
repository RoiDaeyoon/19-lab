# 주제 : 아크릴산 제조공정

 아크릴산(Crude Acrylic Acid·CAA)은 주로 NCC(나프타 분해공정 중 프로필렌 생산)$\rightarrow$ 아크릴산$\rightarrow$ SAP의 단계로써 프로필렌을 고온에서 산소와 반응시켜 생성되는 액상 제품이며, 아크릴산은 **SAP(고흡수성 수지·Super Absorbent Polymer)·도료·점착제 등 다양한 용도로 사용되어 높은 시장가치를 보인다.


<BR> 아크릴산 사업은 국내 **LG화학**을 비롯 독일 바스프(BASF), 미국 다우(DOW), 일본 미쯔비시 등 세계적인 기업만 고유의 공정기술을 갖추고 있다.
<br>또한, LG화학은 아크릴산 촉매 및 제조공정까지 독자기술로 개발했다..

**SAP : 자기 무게의 최대 500배에 달하는 물을 흡수할 만큼 뛰어난 흡수력과 보수력(압력을 가해도 흡수된 물이 빠져나가지 않는 특성)을 가진 고분자 소재. 기저귀 및 위생용품의 핵심 소재로 사용된다.

###제조 공정 소개

 현재 아크릴산을 생산하는 방법으로는 화석연료로부터 얻어지는 프로필렌으로부터 아
크롤레인을 거쳐 아크릴산을 생산하는 공정이 일반적이다.

 한편, **식물유로부터** 글리세롤을 얻어 알릴 알콜로부터 **아크릴산을 생산**하는 경우, 화석 연료의 사용 없이 **친환경적**인 바이오매스로부터 아크릴산
을 생산할 수 있는 **장점**이 있으나, 이 경우 **3-하이드록시프로피오닉산(3-HPA)이 동시에 생성**되어 아크릴산의 **수율이 낮은 문제점**이 있다.

 따라서,**알릴알콜의 산화 반응 생성물**로부터 3-하이드록시프로피오닉산(**3-HPA**)을 효율적으로 **분리**한 뒤 **탈수 반응** 을 통해 **아크릴산으로 변환**시켜야 전체적으로 고수율의 아크릴산을 얻을 수 있어 경제성을 확보할 수 있다.

 따라서 두번째와 같은 생산 공정을 사용한다.

##제조 공정

알릴알콜의 산화반응의 결과로 발생한 아크릴산과  3-하이드록시프로피오닉산(3-HPA) 및 기타 불순물을 포함하는 수용액을 더욱 정제하여 높은 수율의 아크
릴산을 제조하는 방법

1.  **아크릴산(AA)** 및 3-하이드록시프로피오닉산(3-hydroxypropionic acid, **3-HPA**)을 포함하는 **수용액(Feed)**에 **유기 추출제**(MIBK,MTBE,n-부탄올 등)를 사용하여 **아크릴산을 포함하는 유기상** 및 **3-하이드록시프로피오닉산(3-HPA)을 포함하는 수상**으로 **분리**(불순물 포함)

2. **3-하이드록시프로피오닉산(3-HPA)을 포함하는 수상**은 **탈수 반응**을 통해 **아크릴산을 포함하는 수상**으로 **전환** $\rightarrow$ **탈수반응** 사용
<br>image.png

3. **아크릴산을 포함하는 유기상** 및 단계2 에서 얻어진 **아크릴산을 포함하는 수상**의 **혼합 용액**에서 **유기 추출제를 분리** 하여 **아크릴산을 수득**하여 제조 $\rightarrow$  **증류방법** 사용

image.png

1. **아크릴산 및 3-HPA 수용액**은 반응물 이송라인(**2**)으로, **물과 유기 추출제**는 이송라인(**1**) 및 이송라인(**3**)을 통해서 **추출장치(100)**로 유입

2. 분리된 **아크릴산을 포함하는 유기상**은 이송라인(**4**)을 통해서 **아크릴산 정제장치(400)**로 유입

3. 상기 추출장치(100)에서 분리된 **3-HPA를 포함하는 수상**은 이송라인(**5**)을 통해서 **3-HPA 정제장치(200)**로 유입되어 **정제**된 후, 이송라인(**7**)을 통해서 **탈수반응장치(300)**로 유입 

4. 상기 **탈수반응장치(300)**에서는 **3-HPA을 포함하는 수상**이 **탈수반응** 처리되어 **아크릴산을 포함하는 수상**으로 전환.

4. **전환된 아크릴산을 포함하는 수상**은 이송라인(**8**)을 통해서 **아크릴산 정제장치(400)**로 이송

5. 이 때, **아크릴산 정제장치(400)**로 **이송되기 전** 상기 이송라인(**4**)과 이송라인(**8**)이 서로 만나, 아크릴산을 포함하는 유기상 및 아크릴산을 포함하는 수상은 **혼합**되어 **아크릴산 정제장치(400)**로 유입

6. 상기 혼합물은 **아크릴산 정제장치(400)**에서 **정제**되어 이송라인(**11**)을 통해서 아크릴산을 수득

7. 아크릴산 정제장치에서 **분리된 유기 추출제**와 **물을 포함하는 용액**은 **유기 추출제 회수 장치(500)**에서 **정제**하여 **유기 추출제**와 **나머지
폐수**로 분리

8. **정제된 유기 추출제**는 이송라인(**12**)을 통해서 **재활용**된다

### 고려사항

1. **단계 1**에서 수용액과 유기추출제의 질량비는 1:5 내지 5:1로 정한다. 

2. **단계 2**의 혼합용액에서 촉매(산성 또는 염기성 촉매), 액상 또는 기상의 상 변화, 반응 온도(70 ~ 300$^{\circ}{\rm C}$) 및 시간 등 고려한다.

3. 아크릴산 수율은 80%이상이 되도록 한다.
